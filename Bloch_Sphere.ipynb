```python
import qutip
import numpy as np

initial_state = np.array((0,1))
final_state = X @ initial_state

print(initial_state)
print(final_state )

b = qutip.Bloch(figsize=(3, 3))

B_initial_state = qutip.Qobj(initial_state)
B_final_state = qutip.Qobj(final_state)

b.add_states([B_initial_state, B_final_state])
b.show()
```

```python
import qutip as qt
import numpy as np

theta = np.pi / 2
n = (0, 1, 1)

R = rotation_gate(theta, n)   

initial_state = np.array([0, 1])
final_state = R @ initial_state

print(initial_state)
print(final_state)

B_initial = qt.Qobj(initial_state)
B_final = qt.Qobj(final_state)

b = qt.Bloch(figsize=(3, 3))
b.add_states([B_initial, B_final])
b.show()
```
