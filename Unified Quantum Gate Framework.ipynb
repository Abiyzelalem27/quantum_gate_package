Case 2: Control = second qubit, Target = first qubit
          CNOT1→0=I⊗P0+X⊗P1

```python
import numpy as np
CNOT= np.kron(I, P0) + np.kron(X, P1)
print(CNOT)

#Example 
initial_state=np.array((1,0,0,0))
final_state=CNOT@initial_state
print(final_state)
```
