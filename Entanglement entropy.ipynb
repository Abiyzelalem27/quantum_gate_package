The general form of a pure entangled state for two qubits, according to the
Schmidt decomposition and up to local unitaries, is given by |ψ⟩ = cos θ |00⟩ +
sin θ |11⟩. Plot its entanglement entropy EE(|ψ⟩) as a function of θ.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter range: theta in [0, pi/2]
theta = np.linspace(0, np.pi/2, 1000)

# Eigenvalues of the reduced density matrix
lam = np.cos(theta)**2

# Small number to avoid log(0)
eps = 1e-15
lam_clipped = np.clip(lam, eps, 1 - eps)

# Entanglement entropy in bits
E = -(lam_clipped*np.log2(lam_clipped) +
      (1 - lam_clipped)*np.log2(1 - lam_clipped))

# Plot
plt.figure(figsize=(7, 4.2))
plt.plot(theta, E)

plt.xlabel(r'$\theta$')
plt.ylabel(r'$E_E(|\psi\rangle)$ (bits)')


plt.xlim(0, np.pi/2)
plt.ylim(0, 1.05)

# MORE angle ticks
plt.xticks(
    [0, np.pi/8, np.pi/4, 3*np.pi/8, np.pi/2],
    [r'$0$', r'$\pi/8$', r'$\pi/4$', r'$3\pi/8$', r'$\pi/2$']
)

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
