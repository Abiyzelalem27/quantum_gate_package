Create a function that generates a complete set of projectors
(computational basis)

{𝑃𝑖}, 𝑃𝑖=∣𝑖⟩⟨𝑖∣

```python
import numpy as np

def projectors(dim):
    """
    Generate computational basis projectors {|i><i|}.
    """
    projectors = []
    for i in range(dim):
        ket = np.zeros(dim)
        ket[i] = 1.0
        P = np.outer(ket, ket)
        projectors.append(P)
    return projectors

    #Example
P = projectors(2)
print(P[0])
print(P[1])
```

```python
import numpy as np

def projector(dim):
    """
    Generate computational basis projector {|i><i|}.
    """
    projector = []
    for i in range(dim):
        ket = np.zeros(dim)
        ket[i] = 1
        P = np.outer(ket, ket)
        projector.append(P)
    return projector

    #Example
P = projector(4)
print(P[0])
print(P[3])
```

Born rule function

𝑝𝑖=Tr(𝑃𝑖𝜌)

```python
import numpy as np

def born_rule(rho, projectors):
    """
    Compute the Born-rule probability vector p_i = Tr(P_i rho).

    Parameters
    ----------
    rho : numpy.ndarray
        Density matrix.
    projectors : list of numpy.ndarray
        Computational basis projectors.

    """
    probs = np.array([np.trace(P @ rho).real for P in projectors])
    probs = probs / np.sum(probs)   # normalize (numerical safety)
    return probs
```

Sample from probability vector

(cumulative sum + random number, as written)

```python
import numpy as np

def sample_probability(probs):
    """
    Sample an index from a probability vector using cumulative sums.
    """
    r = np.random.rand()
    cumulative = 0.0

    for i, p in enumerate(probs):
        cumulative += p
        if r < cumulative:
            return i

    return len(probs) - 1  # fallback (numerical)
```

Define measurement(ρ)

```python
import numpy as np

def measurement(state):
    """
    Perform a computational-basis measurement.

    Parameters
    ----------
    state : numpy.ndarray
        State vector or density matrix.

    """
    
    if state.ndim == 1:
        rho = np.outer(state, state.conj())
    elif state.ndim == 2:
        rho = state
    else:
        raise ValueError("State must be vector or density matrix")

    dim = rho.shape[0]
   P = projectors(dim)
    probs = born_rule(rho, P)

    outcome = sample_probability(probs)
    return outcome

    psi = np.array([1/np.sqrt(2), 1/np.sqrt(2)])
print(measurement(psi))
```
